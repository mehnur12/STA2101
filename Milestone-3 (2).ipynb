```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Healthcare Dataset: Frequency and Graphical Analysis")
print("Name: M.H. Ansha Hossain")
print("ID: 242016008")
print("Section: 04\n")

print("------------------------------------------------------------")
print("Frequency Distribution Table")
print("------------------------------------------------------------\n")

data = {
    "Classes/Categories": [
        "Arthritis", "Asthma", "Cancer", "Diabetes",
        "Heart Disease", "Kidney Disease", "Obesity",
        "Stroke", "Others"
    ],
    "f": [14, 10, 18, 12, 8, 6, 11, 5, 4]
}

df = pd.DataFrame(data)
N = df["f"].sum()
df["rf"] = np.round(df["f"] / N, 4)
df["cf"] = df["f"].cumsum()
df["rcf"] = np.round(df["cf"] / N, 4)

print("Complete Frequency Distribution of Patient Medical Conditions (N = {})\n".format(N))
print(df.to_string(index=False))
print()

print("------------------------------------------------------------")
print("Graphical Analysis")
print("------------------------------------------------------------\n")

plt.figure(figsize=(8, 5))
plt.bar(df["Classes/Categories"], df["f"], color='skyblue', edgecolor='black')
plt.title("Frequency of Patient Medical Conditions")
plt.xlabel("Medical Conditions")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(6, 6))
plt.pie(df["rf"], labels=df["Classes/Categories"], autopct='%1.1f%%', startangle=140)
plt.title("Relative Frequency Distribution of Conditions")
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(df["Classes/Categories"], df["cf"], marker='o', linestyle='-', color='green')
plt.title("Cumulative Frequency Distribution")
plt.xlabel("Medical Conditions")
plt.ylabel("Cumulative Frequency")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.show()

print("------------------------------------------------------------")
print("Analysis and Interpretation")
print("------------------------------------------------------------\n")

print("From the analysis, it is observed that Cancer has the highest frequency among the given medical conditions,")
print("indicating a relatively higher occurrence rate in the dataset. Arthritis and Diabetes also show significant frequencies,")
print("while conditions like Stroke and Others appear less frequently. The cumulative and relative frequencies provide a clear understanding")
print("of how each condition contributes to the total dataset distribution.\n")

print("------------------------------------------------------------")
print("Challenges Faced")
print("------------------------------------------------------------\n")

print("Several challenges were faced while writing the code in Overleaf. Difficulties were encountered in plotting the frequency graph and other charts,")
print("as minor syntax errors often caused compilation issues. Proper alignment and scaling of the figures were hard to achieve initially,")
print("and adjustments had to be made repeatedly until the desired output was obtained. Managing the placement of graphs, text, and tables")
print("also required careful formatting and multiple revisions before the final layout was achieved.\n")

print("------------------------------------------------------------")
print("Conclusion")
print("------------------------------------------------------------\n")

print("The study successfully demonstrates the use of data analysis techniques to represent healthcare information in both tabular and graphical forms.")
print("By using pandas and numpy for computation and matplotlib for visualization, the dataset was effectively analyzed to show the distribution")
print("and relationship among different medical conditions. This process helped in better understanding the dataset and interpreting its key patterns.")

```
